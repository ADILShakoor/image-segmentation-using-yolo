```python
import numpy as np

# Function to calculate the area of a single mask
def calculate_area(mask):
    if isinstance(mask, np.ndarray):
        return np.sum(mask)
    else:
        # Convert mask to a NumPy array if it's not already
        try:
            flat_mask = np.concatenate([np.array(m) for m in mask], axis=None)
            return np.sum(flat_mask)
        except Exception as e:
            print("Error processing mask:", e)
            return 0  # Return 0 if there's an error

# Assuming you already have the results from the prediction
results = model.predict("prediction_img/rotten01.jpg", save=True, save_txt=True, imgsz=640, conf=0.5)

# Extract masks and class labels
masks = results[0].masks  # Get mask data
class_ids = results[0].boxes.cls  # Get class IDs

# Initialize areas
rotten_apple_area = 0
rotten_part_area = 0

# Loop through each mask and corresponding class ID
for i in range(len(masks)):
    mask = masks[i].data.numpy()  # Convert to a NumPy array if necessary
    class_id = int(class_ids[i])

    # Check if the class ID corresponds to rotten_apple or rotten_part
    if class_id == 1:  # Class ID for rotten_apple
        rotten_apple_area += calculate_area(mask)
    elif class_id == 2:  # Class ID for rotten_part
        rotten_part_area += calculate_area(mask)

# Step 3: Calculate the percentage of the rotten part
if rotten_apple_area > 0:
    if rotten_part_area > 0:
       rotten_percentage = (rotten_part_area / rotten_apple_area) * 100
       print(f"Rotten part percentage: {rotten_percentage:.2f}%")
       if rotten_percentage <50:
           freshness="Fresh"
    else:
        print(f"Rotten part percentage: 98%")
        freshness="Rotten"
else:
    print(f"Rotten  percentage: 0%")
    freshness="Fresh"
    # print("No rotten apple detected.")
print(f"freshness: {freshness} ")
```
